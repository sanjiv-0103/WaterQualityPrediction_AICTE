```python
#initializing and importing the files
import pandas as pd
import numpy as np

from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#Reading the given csv file
df = pd.read_csv('PB_All_2000_2021.csv',sep=';')
df
```

```python
#Structure of the DataFrame
df.info()
```

```python
#Dimensions of the table- Rows X Columns
df.shape
```

```python
#Transposing the table
df.describe().T
```

```python
#Count the number of missing values in each column
df.isnull().sum()
```

```python
#Changing the date format
df['date']=pd.to_datetime(df['date'],format='%d.%m.%Y')
df
```

```python
 df.info()
```

```python
#Sorting the values
df=df.sort_values(by=['id','date'])
#Top 5 Row informations 
df.head()
```

```python
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month
```

```python
df.head()
```

```python
#View and access column names
df.columns
```

```python
pollutants=['O2', 'NO3', 'NO2', 'SO4',
       'PO4', 'CL']
```

### WEEK 2

```python
df=df.dropna(subset=pollutants)
df.head()
```

```python
df.isnull().sum()
```

```python
X=df[['id','year']]
Y=df[pollutants]
```

```python
X_encoded=pd.get_dummies(X,columns=['id'],drop_first=True)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(
    X_encoded, Y, test_size=0.2, random_state=42
)
```

```python
model=MultiOutputRegressor(RandomForestRegressor(n_estimators=100,random_state=42))
model.fit(X_train, Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('      MSE:',mean_squared_error(Y_test.iloc[:, i], Y_pred[:, i]))
    print('      R2:',r2_score(Y_test.iloc[:, i], Y_pred[:, i]))
    print()
```

```python
station_id = '22'
year_input = 2024

input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
input_encoded = pd.get_dummies(input_data, columns=['id'])

missing_cols = set(X_encoded.columns) - set (input_encoded.columns) 
for col in missing_cols:
    input_encoded[col] = 0
input_encoded = input_encoded [X_encoded.columns]

predicted_pollutants=model.predict(input_encoded)[0]

print(f"InPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f" {p}: {val:.2f}")
```
